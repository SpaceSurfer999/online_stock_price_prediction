```python
import yfinance as yf
import mercury as mr
import mplfinance as mpf
from datetime import date, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mdate
```

```python
app = mr.App(title="Parameters", description="Report financial data for selected ticker")
```

```python
ticker_name = mr.Text(label="What stock do you want to analyze?", value="AAPL")
```

```python
tf_list = ['1h', '1d', '5d', '1wk', '1mo']
```

```python
 tf = mr.Select(label="Select timeframe", value="1d", choices=tf_list)
```

```python
data_slider = mr.Slider(label='Select period', value=100, min=5, max=500)
```

```python
# weeks = mr.Slider(label="Weeks", value=52, min=4, max=260)
```

```python
# start = date.today() - timedelta(weeks.value*7)
# end = date.today() + timedelta(2)
# df = pd.DataFrame(yf.download(ticker.value, start=start, end=end, progress=False))
```

```python
start = date.today() - timedelta(data_slider.value)
```

```python
history = yf.Ticker(ticker_name.value).history(interval=tf.value, start=start)
```

```python
mpf.plot(history, type='candle', mav=(21),figratio=(21,10))
```

```python
mr.Md(f"""## Basic statistics
#### Local low: {history.Low.min().round(2)}
#### Local high: {history.High.max().round(2)}""")
```
